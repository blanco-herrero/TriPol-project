# Media

```python
#get a list with all the paths of all files
root = "/Users/davidblancoherrero/Documentos/Proyecto tripol/DESCARGA_MEDIOS" # it may have many subfolders and files inside
lst = []
from fnmatch import fnmatch
pattern = "*.txt"      #I want to get only txt files 
#pattern = "*.*"        # Note: Use this pattern to get all types of files and folders 
for path, subdirs, files in os.walk(root):
    for name in files:
        if fnmatch(name, pattern):
            lst.append((os.path.join(path, name)))
files = pd.DataFrame({"filePaths":lst})
files.to_csv("filepaths.csv")
files['filePaths']
#files = files.values.tolist()

files
```

```python
#Read file from line 2 and split into paragraphs to a list

def transform_to_paragraphs (filename):
    

    with open(filename) as f:
        country=f.name[f.name.rindex('/')+1:f.name.rindex('/')+3]
        reference=f.name[f.name.rindex('/')+1:]
        lines = f.read()

        tests = lines.split("\n\n")
        match = re.search(r'\d{2}/\d{2}/\d{4}', tests[0])
        try:
            date = datetime.datetime.strptime(match.group(), '%d/%m/%Y').date() 
        except AttributeError:
            return "NaN"
        except ValueError:
            return "NaN"
        try:
            media = tests[0].split(". ",1)[1]
        except IndexError:
            return "NaN"

        df = pd.DataFrame(tests[2:])
        df.rename(columns={0:'text'}, inplace=True)
        df['date']=date
        df['media']=media
        df['country']=country
        df['reference']=reference
        df.to_csv(r'data/'+f'{reference}.csv')
```

```python
for i in files['filePaths']:
    transform_to_paragraphs(i)
```

```python
def prepend(list, str):
    str += '{0}'
    list = [str.format(i) for i in list]
    return(list)

filepaths = [f for f in listdir("./data") if f.endswith('.csv')]
str = 'data/'
filepaths = prepend(filepaths, str)
df = pd.concat(map(pd.read_csv, filepaths))

print("Number of files: ", len(filepaths))
print("Number of records: ", len(df))
```
